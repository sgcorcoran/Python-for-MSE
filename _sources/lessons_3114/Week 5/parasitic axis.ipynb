```python
import numpy as np
```

```python
data=np.array([[ 5.12012079e+01,  9.98176335e+01],
       [ 3.65137520e+01,  9.51043593e+01],
       [ 3.12191540e+01,  8.48428362e+01],
       [ 2.77593947e+01,  7.45955668e+01],
       [ 2.49896435e+01,  6.53756969e+01],
       [ 2.26756865e+01,  5.46263088e+01],
       [ 2.03627013e+01,  4.43879480e+01],
       [ 1.80506880e+01,  3.46606143e+01],
       [ 1.55083480e+01,  2.44204717e+01],
       [ 1.27376250e+01,  1.46895746e+01],
       [ 8.36044633e+00,  4.43517838e+00],
       [ 4.68299427e+00,  3.18453503e-01],
       [ 9.49257746e-02, 2.28207909e-01]])
```

```python
np.log10(-np.log(1-data[:,1]/100))
```

```python
from mpl_toolkits.axes_grid1 import host_subplot
from mpl_toolkits import axisartist
import matplotlib.pyplot as plt
import numpy as np

host = host_subplot(111, axes_class=axisartist.Axes)
plt.subplots_adjust(right=0.75) #size of plot?

par1 = host.twinx() 
par2 = host.twinx()

par2.axis["right"] = par2.new_fixed_axis(loc="right", offset=(60, 0))

par1.axis["right"].toggle(all=True)
par2.axis["right"].toggle(all=True)

p1, = host.plot(data[:,0], data[:,1], label="Density")
p2, = par1.plot(data[:,0], np.log10(-np.log(1-data[:,1]/100)), label="Log data 1")
p3, = par2.plot(data[:,0], data[:,1], label="data 1")

par2.set_yscale('log')

host.set_xlim(0, 60)
host.set_ylim(0, 100)
par1.set_ylim(0, 2)
par2.set_ylim(1, 100)

host.set_xlabel("Distance")
host.set_ylabel("Density")
par1.set_ylabel("Log data 1")
par2.set_ylabel("data 1")

host.legend()

host.axis["left"].label.set_color(p1.get_color())
par1.axis["right"].label.set_color(p2.get_color())
par2.axis["right"].label.set_color(p3.get_color())

plt.show()
```
