```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xdata=np.linspace(0.005, 0.1, 1000)
xdata
ydata= np.e**(0.06/xdata) #this has the form y=e^(k/x)
#Note:  ln(y) = k/x  or log(y)=k/(2.3 x)

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3,figsize=(12, 6))

ax1.scatter(1/xdata, ydata, marker='.', color='b')
ax1.set_xlabel('1/xdata')
ax1.set_ylabel('ydata')
ax2.semilogy(1/xdata, ydata, 'b.')
ax2.set_xlabel('1/xdata')
ax2.set_ylabel('ydata')
ax3.scatter(1/xdata, np.log10(ydata), marker='.', color='r')
ax3.set_xlabel('1/xdata')
ax3.set_ylabel('log10(ydata)')

ax2.tick_params(axis='y',which='minor', length=0)

fig.tight_layout()
```

```python
fig, (ax2, ax3) = plt.subplots(nrows=1, ncols=2,figsize=(12, 6))

ax2.semilogy(1/xdata, ydata, 'b.')
ax2.set_ylim(100, 1000)
ax2.set_xlabel('1/xdata')
ax2.set_ylabel('ydata')
ax2.set_xlim(70, 120)
ax2.tick_params(axis="y", which='minor',labelleft=False)
ax2.tick_params(axis="y", which='major',labelsize=12)
ax3.scatter(1/xdata, np.log10(ydata), marker='.', color='r')
ax3.set_xlabel('1/xdata')
ax3.set_ylabel('log10(ydata)')
ax3.set_ylim(2, 3)
ax3.set_xlim(70, 120)

ax2.tick_params(axis='y',which='minor', length=5)

fig.tight_layout()
```
